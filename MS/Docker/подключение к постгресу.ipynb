```python
import psycopg2
import pandas as pd
from datetime import datetime


conn = psycopg2.connect(
    host="db", 
    database="polina_nch",
    user="polina",
    password="nesterovich"
)
cur = conn.cursor()

print("Подключение к БД настроено")
```

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS customers (
    id SERIAL PRIMARY KEY,
    customer_name TEXT,
    age INT,
    city TEXT,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
""")
conn.commit()
```

```python
customers_data = [
    ("Иванов Иван", 30, "Москва"),
    ("Зыкова Надежда", 55, "Санкт-Петербург"),
    ("Воронов Никита", 23, "Иркутск")
]

for customer in customers_data:
    cur.execute(
        "INSERT INTO customers (customer_name, age, city) VALUES (%s, %s, %s)",
        customer
    )
conn.commit()
```

```python
cur.execute("SELECT * FROM customers")
rows = cur.fetchall()
column_names = [desc[0] for desc in cur.description]

df = pd.DataFrame(rows, columns=column_names)
df
```
