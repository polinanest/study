```scala
%spark.pyspark
from pyspark.sql import functions as F
users = spark.createDataFrame(
    [
        ("u1", "Berlin"),
        ("u2", "Berlin"),
        ("u3", "Munich"),
        ("u4", "Hamburg"),
    ],
    ["user_id", "city"]
)
orders = spark.createDataFrame(
    [
        ("o1", "u1", "p1", 2, 10.0),
        ("o2", "u1", "p2", 1, 30.0),
        ("o3", "u2", "p1", 1, 10.0),
        ("o4", "u2", "p3", 5, 7.0),
        ("o5", "u3", "p2", 3, 30.0),
        ("o6", "u3", "p3", 1, 7.0),
        ("o7", "u4", "p1", 10, 10.0),
    ],
    ["order_id", "user_id", "product_id", "qty", "price"]
)
products = spark.createDataFrame(
    [
        ("p1", "Ring VOLA"),
        ("p2", "Ring POROG"),
        ("p3", "Ring TISHINA"),
    ],
    ["product_id", "product_name"]
)
users.show()
orders.show()
products.show()
```

```scala
%spark.pyspark
orders = orders.withColumn("revenue", F.col("qty") * F.col("price"))

orders.show()
```

```scala
%spark.pyspark
df = orders.join(users, "user_id").join(products, "product_id")
df.show()
```

```scala
%spark.pyspark
metrics = df.groupBy("city", "product_id", "product_name").agg(
    F.count("*").alias("orders_cnt"),
    F.sum("qty").alias("qty_sum"),
    F.sum("revenue").alias("revenue_sum")
)

metrics.show()
```

```scala
%spark.pyspark
output_path = "hdfs:///tmp/sandbox_zeppelin/mart_city_top_products/"
top_products.write.mode("overwrite").parquet(output_path)
```
